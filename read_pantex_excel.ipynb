```python
import pandas as pd
```

```python

df= pd.read_excel('CXRay.xlsx')
# print(df.columns)
# Coose just the relevant columns
df=df[['PFW_ID','ExamDate','ReadingDate','ParAbn','PleuAbn','OtherAbn','OtherFindings']]
# Filter out any studies with abnormal findings
df = df[ 
  (df.ParAbn == 'N') & 
  (df.PleuAbn == 'N') & 
  (df.OtherAbn == 'N') & 
  (df.OtherFindings.isna()) ]
# Make the exam date a sortable format
df["ExamDate"] = pd.to_datetime(df["ExamDate"])
# Sort by exam date
df = df.sort_values(by='ExamDate', ascending = False)
#get first 40 secords
df.head(40)
```
